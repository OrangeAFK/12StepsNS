```python
# we go from 1-D linear convection to non-linear simply by changing the speed term from a constant to u

import numpy as np
from matplotlib import pyplot
%matplotlib inline

cells = 41
dx = 2 / (cells + 1)
timesteps = 20
dt = .025
c=1

u = np.ones(cells)
u[int(.5/dx):int(1/dx + 1)] = 2

for t in range(timesteps):
  u[1:] -= (u[1:]*dt/dx)*(u[1:] - u[:-1])

pyplot.plot(np.linspace(0, 2, cells), u)
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.title('Nonlinear convection')
```
