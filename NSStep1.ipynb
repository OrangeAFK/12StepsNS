```python
import numpy
from matplotlib import pyplot
import time
import sys
```

```python
%matplotlib inline
```

```python
# modeling the basic convection equation du/dt + c * du/dx = 0

cells = 41
width = 2
dx = width / (cells - 1)
timesteps = 25
dt = timesteps / 1000
c = 1 # wavespeed
```

```python
# set initial conditions: u_0 = 2 if 0.5<=x<=1 else 1
u = numpy.ones(cells)
u[int(0.5/dx) : int(1/dx + 1)] = 2
u
```

```python
pyplot.plot(numpy.linspace(0, 2, cells), u)
```

```python
# apply FTBS scheme to solve for the value of u at each time step
for n in range(timesteps):
  u_next = u.copy()
  u_next[1:] = u[1:] - (c * dt / dx) * (u[1:] - u[:-1])
  u = u_next.copy()
```

```python
pyplot.plot(numpy.linspace(0, 2, cells), u_next)
```
